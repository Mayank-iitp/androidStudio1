```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/fkenews.csv', encoding='latin1')
```

```python
df.head()
```

```python
df= df.drop(["Image","Date","id","Category"],axis=1)
```

```python
df.head()
```

```python
df["Label"].dtype
```

```python
df.tail()
```

```python
df["Label"].value_counts()
```

```python
from sklearn.utils import resample
```

```python
true= df[df["Label"]==0]
false= df[df["Label"]==1]
```

```python
num_sample=len(false)
undersample= resample(true, replace= False, n_samples= num_sample, random_state= 9)
```

```python
df= pd.concat([undersample,false])
```

```python
df.head()
```

```python
df.reset_index(drop= True,inplace= True)
```

```python
df.head()
```

```python
df["content"]= df['Statement']+" "+df["Web"]
```

```python
X= df.drop(["Web","Label","Statement"],axis=1)
```

```python
df.head()
```

```python
X.head()
```

```python
Y= df["Label"]
```

```python
Y.head()
```

```python
Y.shape
```

```python
Y.head()
Y.value_counts()
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size= 10000
```

```python
mess= X.copy()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
mess["content"][1]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(mess)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', mess["content"][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)

```

```python
from tensorflow.keras.layers import Dropout

embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=49)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
import numpy as np

y_pred_binary = np.where(y_pred > 0.5, 1, 0)
confusion_matrix(y_test, y_pred_binary)
```

```python
from sklearn.metrics import accuracy_score
y_pred_binary = np.where(y_pred > 0.5, 1, 0)
accuracy_score(y_test,y_pred_binary)
```

```python
def preprocess_text(text):
  vocab_size= 10000
  encod_text= [one_hot(text, vocab_size)]
  max_length=20
  padded_text= pad_sequences(encod_text,padding='pre',maxlen=max_length)
  return padded_text

def predict_news(text):
  processed_text= preprocess_text(text)
  prediction= model.predict(processed_text)
  is_fake=(prediction > 0.5).astype(int)

  return bool(is_fake[0])
```

```python
text= " india can never overtake china"
isk= predict_news(text)
if isk:
  print("The news is likely to be fake")
else:
  print ("The news is likely to be real")
```
